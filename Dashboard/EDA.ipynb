```python
import re
import numpy as np

def convert_to_int(value):
    if pd.isna(value):
        return np.nan

    num_dict = {'K': 1e3, 'k': 1e3, 'M': 1e6, 'B': 1e9}
    match = re.match(r'^(\d+\.?\d*)([kKMB])$', value)

    if match:
        number, magnitude = match.groups()
        return int(float(number) * num_dict[magnitude])
    else:
        return int(value)
```

```python
def filter_years_columns(dataframe, start_year, end_year):
    # extract the columns that represent years
    year_columns = ['country'] + [str(year) for year in range(start_year, end_year + 1)]

    # filter columns based on the specified range of years
    filtered_data = dataframe[year_columns]

    return filtered_data
```

```python
start_year = 2000
end_year = 2010

filtered_child_mortality = filter_years_columns(child_mortality, start_year, end_year)
filtered_income = filter_years_columns(income, start_year, end_year)
filtered_gdp_per_capita = filter_years_columns(gdp_per_capita, start_year, end_year)
filtered_gini = filter_years_columns(gini, start_year, end_year)
filtered_government_health_spend = filter_years_columns(government_health_spend, start_year, end_year)
filtered_life_expectancy = filter_years_columns(life_expectancy, start_year, end_year)
filtered_num_people_poverty = filter_years_columns(num_people_poverty, start_year, end_year)
filtered_population = filter_years_columns(population, start_year, end_year)
filtered_private_share = filter_years_columns(private_share, start_year, end_year)
filtered_doctors = filter_years_columns(doctors, start_year, end_year)
```
